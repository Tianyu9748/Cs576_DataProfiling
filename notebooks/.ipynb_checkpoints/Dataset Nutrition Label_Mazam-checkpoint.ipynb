## Load Dataset

```python
Datasets = ['adult.csv'] #Place your CSV file here
Protected_Attrs = ['age', 'sex', 'race'] ## Place protected attributes
df = pd.read_csv(Datasets[0]) ## Reading the csv file
```

# Correlations

```python
## using pandas get_numeric_data, we are getting column names which contains numemric values.
selected_cols = list(df._get_numeric_data()) 
## we are creating a seperate DataFrame from selected columns above.
data = df[selected_cols]
## getting column names of the newly created DataFrame
data_c = data.columns
## using corr() function of pandas library we are getting a matrix of correlated values of every columns in out
## new dataframe(data)
data_v = data.corr()
```

### Ploting Heatmap

```python
## Using Seaborn library we plot correlation Heatmap from data_v(correlated matrix that we got above cell.)
sns.heatmap(data_v, annot = True, vmin=-1, vmax=1, center= 0, cmap= 'coolwarm', linewidths=2, linecolor='white', fmt='.2g')
```

### Plotting Relational Graph Matrix

```python
data_num = len(data_c) ## get number of columns in the dataset
## iunitilizing subplots where columns and row numbers are data_num above
fig, axes = plt.subplots(nrows=data_num, ncols=data_num, figsize=(20, 25)) 
## iterate through every row and evey column of subplots
for ni, n in enumerate(data_c):  #row index
    for pi, p in enumerate(data_c):  # column index
        # Call `scatter` or any plot function on the 
        # respective `axes` object itself
        axes[ni,pi].scatter(data[n], data[p])
        ## setting the plot title column names.
        axes[ni,pi].set_title(n+' vs '+p)
```

# Functional Dependencies

```python
from FDTool.fdtool import fdtool ## importing our own FDTool class
import networkx as nx ## importing networkx library
from pyvis.network import Network ## importing Network class from pyvis library
```

```python
FD, ASSC = fdtool.main(Datasets[0], 100) #<--- It takes time(Change the number to larger number.)(Number of smaples=100)
```

```python
lGraph = [] ## Empty list
for fd in FD: # iterate through evry subset in the list FD
    if len(fd[1])>=3: ##Checking if subset length is greater than 3
        if fd[0] in Protected_Attrs: # cheking the sbuset contains our pretected attributes
            lGraph.append((fd[0], fd[1])) # appeding the subset into the list created above
```

```python
## Setting the matplotlib in notebook setting
%matplotlib notebook

## Initilizing Network Class from pyvis.
g = Network(height=900, width=900, notebook=True)

h = [] ## Empty list
for lg in lGraph: ## itarate through list from above Cell.
    if str(lg[0]) in h and str(lg[1]) in h: ## Checking if h contains any item from the list
        continue
    g.add_node(str(lg[0]), title=str(lg[0]), size=5) ## Adding the first element from item as a node in the Netowrk
    g.add_node(str(lg[1]), title=str(lg[1]),size=5) ## Adding the first element from item as a node in the Netowrk
    g.add_edge(str(lg[0]), str(lg[1])) ## Adding both element of item as a Edge of the network
    ## Appending h with both element of the item.
    h.append(str(lg[0]))
    h.append(str(lg[1]))
## Showing the network in html format.
g.prep_notebook()
g.show('ex.html')
```

# Association Rules

```python
assGraph = [] ## Empty LIst
for assc in ASSC: ## iterate through every item of ASSC list.
    if len(assc[1])>=3: ## Checking each subset length is greater than 3.
        assGraph.append((assc[0], assc[1])) ## Appending the list with both element of the subset.
```

```python
## Initilizing Network Class from pyvis.
g = Network(height=900, width=900, notebook=True)
# g.barnes_hut()
h = []
for lg in assGraph:## itarate through list from above Cell.
    if str(lg[0]) in h and str(lg[1]) in h:## Checking if h contains any item from the list
        continue
    g.add_node(str(lg[0]), title=str(lg[0]), size=5) ## Adding the first element from item as a node in the Netowrk
    g.add_node(str(lg[1]), title=str(lg[1]),size=5)## Adding the first element from item as a node in the Netowrk
    g.add_edge(str(lg[0]), str(lg[1]))## Adding both element of item as a Edge of the network
    ## Appending h with both element of the item.
    h.append(str(lg[0]))
    h.append(str(lg[1]))
## Showing the network in html format.
g.prep_notebook()
g.show('ex1.html')
```
